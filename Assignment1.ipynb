Task A

```python
import random

N = 500
adj = [[] for _ in range(N)]
edges = 0

random.seed()

for i in range(N):
    deg = random.randint(2, 7)  # rand()%6 + 2 -> 2..7

    while len(adj[i]) < deg:
        j = random.randint(0, N - 1)

        if i == j:
            continue

        present = any(x[0] == j for x in adj[i])

        if present:
            continue

        w = random.randint(5, 50)  # rand()%46 + 5 -> 5..50

        adj[i].append((j, w))
        adj[j].append((i, w))

        edges += 1

print("Total Cities =", N)
print("Total Roads =", edges)
```

Task B

```python
import sys
sys.setrecursionlimit(10000)

vis = [False] * N
component = []

def dfs(u):
    vis[u] = True
    component.append(u)

    for x in adj[u]:
        v = x[0]
        if not vis[v]:
            dfs(v)

largest = []

for i in range(N):
    if not vis[i]:
        component.clear()
        dfs(i)

        if len(component) > len(largest):
            largest = component.copy()

print("Largest Component Size =", len(largest))
print(*largest)
```

Task C

```python
import sys
sys.setrecursionlimit(10000)

vis = [False] * N
component = []

def dfs(u):
    vis[u] = True
    component.append(u)

    for x in adj[u]:
        v = x[0]
        if not vis[v]:
            dfs(v)

sizes = []

for i in range(N):
    if not vis[i]:
        component.clear()
        dfs(i)
        sizes.append(len(component))

print("Total Components =", len(sizes))
print(*sizes)

print("Largest Cluster =", max(sizes))
print("Smallest Cluster =", min(sizes))
```

Task D

```python
import heapq

source = 0

dist = [float("inf")] * N

pq = [(0, source)]
dist[source] = 0

while pq:
    d, u = heapq.heappop(pq)

    if d > dist[u]:
        continue

    for x in adj[u]:
        v = x[0]
        w = x[1]

        if dist[v] > dist[u] + w:
            dist[v] = dist[u] + w
            heapq.heappush(pq, (dist[v], v))

mx = -1
city = -1

for i in range(N):
    if dist[i] == float("inf"):
        print(f"City {i} : Unreachable")
    else:
        print(f"City {i} : {dist[i]}")

        if dist[i] > mx:
            mx = dist[i]
            city = i

print("Maximum Distance City =", city)
print("Distance =", mx)
```
